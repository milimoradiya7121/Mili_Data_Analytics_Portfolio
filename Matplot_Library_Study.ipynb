# Matplot Library Features:
- Core Functionality: It provides tools to create a wide variety of plots, including line plots, scatter plots, bar charts, histograms, and more.
- Highly customizable, allowing users to control every aspect of a plot, from colors and line styles to axes and labels.
- Integrates well with other libraries such as NumPy, Pandas, and SciPy, making it a versatile tool in the data analysis and scientific computing stack.
- Simple to use for basic plots with its pyplot module, which mimics MATLAB-like plotting functions.
- Offers advanced features through its object-oriented API, enabling the creation of complex and sophisticated plots.

```python
import matplotlib.pyplot as plt
%matplotlib inline
#helps to see plots in notebook
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
print(x)
print(y)
```

```python
#FUNCTIONAL
plt.plot(x,y,'r')
#plt.show()    using this does not show memory space line
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
plt.subplot(1,2,1)   #rows,cols,no.of plots
plt.plot(x,y,'r',linestyle='--',marker='*',markersize=10,markerfacecolor='green')
#also used as ls
#markeregdewidth, markeredgecolor
plt.subplot(1,2,2)
plt.plot(x,y,'g',linewidth=3) #also used as lw
```

```python
#object-oriented method
fig=plt.figure()  #created blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height of the axis
#values between 0 to 1 and are in percentages
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])  
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Big')
axes2.set_title('Small')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
#will only plot blank graphs

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label = 'X cubed')
ax.legend(loc=(0.6,0.2))  #loc wrt to x-axis
```

```python
import numpy as np
x=np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig = plt.figure()
ax3 = fig.add_axes([0,0,1,1])
ax4 = fig.add_axes([0.2,0.5,.4,.4])
ax3.plot(x,z)
ax3.set_xlabel('X')
ax4.plot(x,y)

ax4.set_xlim([20,22])
ax4.set_ylim([30,50])
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axes[0].plot(x,y,linewidth=3)
axes[1].plot(x,z,linestyle='--')
```
